```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("Student performance.csv")
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="parental level of education", data=df, palette="viridis", order=df["parental level of education"].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel("Parental Education Level")
plt.ylabel("Number of Students")
plt.title("Number of Students by Parental Education Level")
plt.show()
```

From the above graph we can tell that the most of the students come from the parents who have high school or associate's degree but only fewer students have parents with master's or bachelor's degree. We can analyze whether parental education affects students performance.

```python
df["lunch"].value_counts().plot.pie(autopct="%1.1f%%", startangle=180)
plt.ylabel("")
plt.title("Lunch Type Distribution")
plt.show()
```

From the above graph we can see the number of students receive free/reduced lunch or standard lunch.

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df["writing score"], fill=True, color="purple")
plt.xlabel("Writing Score")
plt.ylabel("Density")
plt.title("Density of Writing Scores")
plt.show()
```

From the above graph we can tell that most students score around the average. But there are less students with low or high writing score.

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="math score", y="writing score", data=df, scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.title("Regression Plot of Math vs. Writing Scores")
plt.show()
```

From the above graph we can see that there is a positive correlation between math score and writing score. That means the students who is performing well in math are also performing well in writing.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="test preparation course", hue="gender", data=df, palette="Set2")
plt.xlabel("Test Preparation Course")
plt.ylabel("Number of Students")
plt.title("Students Who Took Test Preparation Course by Gender")
plt.show()
```

From the above graph we can see that there are more females that took the test preparation course than males.
